### Dataset Source <a id=dataset_source></a>

The dataset used in this report was sourced from Kaggle at https://www.kaggle.com/rajeevw/ufcdata (Warrier, 2019). The dataset includes all events from the dawn of the Ultimate Fighting Championship (UFC) in 1993 to 21 March 2021. 

### Dataset Details <a id=dataset_details></a>

Each row in the dataset represents a fight and contains both fighters' statistics leading up to the match. Fighters are labeled red and blue. Each fighter's statistics include average damage, inflicted and received (denoted by the prefix `'opp'`), in all of his previous fights; detailed fight records including winning and losing streaks; and such biometrics as height, reach and age. The overall structure of the dataset as it stands now is most appropriate for predictive modeling, but ill-fitted for plotting for insights. Hence we will modify the dataset to plot more meaningful plots.

The raw dataset has a total of 144 columns and 6013 rows. We shall reduce the number of features and drop observations with missing values  in the upcoming phase of data cleaning and pre-processing.

### Response Variable <a id='response_variable'></a>

When it comes to sports, predicting the winner is generally of most interest among regular viewers, ardent sports fans and sports data analysts. However, since our target feature should be numerical, not categorical, we have decided to create a derived feature `R_post_fight_win_streak` from existing `R_current_win_streak` and `Winner`. 

`R_current_win_streak` does not accurately reflect the winning streak in all the previous fights before the current fight, but it holds the same value as `R_longest_win_streak`. First, we will rectify the values of the former feature. Next, the result of the current fight, represented in `Winner`, will be factored in so that the target feature is set to 0 or `R_current_win_streak` $+$ 1, depending on the outcome of a fight.

## Data Cleaning and Preprocessing <a id='data_cleaning_and_preprocessing'></a>

### Overview

In this section, we will take prerequisite steps to cleaning our data for predictive modelling. The preprocessing steps are as follows: 

1. Check whether there is any oddity in the values of features.

2. Change names of columns and values as required

3. Address irrelevant features.

4. Check for rows with missing or unusual values and drop them.

5. Create another version of `df` without single outliers.

We will check whether our features `_current_win_streak` (throughout the report, the absence of the prefix *R* or *B* at the beginning of a column's name refers to both fighters' features), store appropriate values in line with the fight history. We will do so by checking fight records of the legendary fighter, Anderson Silva. 

```python
df.loc[(df['R_fighter']== 'Anderson Silva') |(df['B_fighter']== 'Anderson Silva'), ['date','Winner', 'B_fighter', 'B_draw', 'B_current_win_streak', 'R_fighter', 'R_draw', 'R_current_win_streak']][11:1:-1]
```

As we can see above, we do not see the expected value of 0 after Silva's defeat to Weidman (index 3492); Also his R_draw is 0 (index 2438) after his draw with Diaz (index 2944). Before we go ahead and fix the column, let us check whether the closely related column `_current_lose_streak` also exhibits the same error. 

```python
df[(df['R_fighter']== 'Anderson Silva') |(df['B_fighter']== 'Anderson Silva')][['date','Winner', 'B_fighter', 'B_draw', 'B_current_lose_streak', 'R_fighter', 'R_draw', 'R_current_lose_streak']][11:1:-1]
```

Despite Silva's consecutive losses against Weidman, we do not see his downfall above. We shall now adjust the dataset so that `_current_win_streak`, `_current_lose_streak` and `_draw` contain accurate information. We will then create our target feature `R_post_fight_win_streak` in the process below. Also, we will create `B_post_fight_win_streak`, `B_post_fight_lose_streak` and `R_post_fight_lose_streak` for plotting purposes only to drop them at the end. 

```python
# Make sure there's no space surroudning the names
df['R_fighter'] = df['R_fighter'].str.strip()
df['B_fighter'] = df['B_fighter'].str.strip()

r_fighters = pd.Index(df['R_fighter'])
b_fighters = pd.Index(df['B_fighter'])

fighters = list(r_fighters | b_fighters)

# Dictionary to hold fighters' names ,winning streaks and draws
win_streaks = {fighter: 0 for fighter in fighters}
lose_streaks = {fighter: 0 for fighter in fighters}
draws = {fighter: 0 for fighter in fighters}

# Traverse from the past to the present
df = df.iloc[::-1]
for index, row in df.iterrows():
    # Update _draw, _current_win_streak and _current_lose_streak so that their values are based on the fight records prior to the current fight
    df.at[index, 'B_draw'] = draws[row['B_fighter']]
    df.at[index, 'R_draw'] = draws[row['R_fighter']]
    df.at[index, 'B_current_win_streak'] = win_streaks[row['B_fighter']]
    df.at[index, 'R_current_win_streak'] = win_streaks[row['R_fighter']]
    df.at[index, 'B_current_lose_streak'] = lose_streaks[row['B_fighter']]
    df.at[index, 'R_current_lose_streak'] = lose_streaks[row['R_fighter']]
    
    if row['Winner'] == 'Red':
        win_streaks[row['R_fighter']] += 1
        win_streaks[row['B_fighter']] = 0 
        lose_streaks[row['B_fighter']] += 1
        lose_streaks[row['R_fighter']] = 0
    elif row['Winner'] == 'Blue':
        win_streaks[row['B_fighter']] += 1
        win_streaks[row['R_fighter']] = 0
        lose_streaks[row['R_fighter']] += 1
        lose_streaks[row['B_fighter']] = 0
    else: # Draw
        win_streaks[row['R_fighter']] = 0
        win_streaks[row['B_fighter']] = 0 
        lose_streaks[row['B_fighter']] = 0
        lose_streaks[row['R_fighter']] = 0
        # If this fight ends up in a draw, store the result in draws
        draws[row['B_fighter']] += 1
        draws[row['R_fighter']] += 1
        
    df.at[index, 'R_post_fight_win_streak'] = win_streaks[row['R_fighter']]    
    df.at[index, 'B_post_fight_win_streak'] = win_streaks[row['B_fighter']]
    df.at[index, 'R_post_fight_lose_streak'] = lose_streaks[row['R_fighter']]    
    df.at[index, 'B_post_fight_lose_streak'] = lose_streaks[row['B_fighter']]    

# Set df in the original order
df = df.iloc[::-1]

df[(df['R_fighter']== 'Anderson Silva') |(df['B_fighter']== 'Anderson Silva')][['date','Winner', 'B_fighter', 'B_draw', 'B_current_win_streak', 'R_fighter', 'R_draw', 'R_current_win_streak', 'R_post_fight_win_streak']][11:1:-1]
                        
                        
```

```python
print(f"B_draw's unique value is {df['B_draw'].unique()}")
print(f"R_draw's unique value is {df['R_draw'].unique()}")
```

Now that we have `_current_win_streak`, `_current_lose_streak` and `_draw` adjusted as needed, we will start cleaning up our dataset. First, we will make all column names lowercase and remove potential whitespace for consistency. We will also rename some columns as required for Phase 2.

Next, since fighters' names themselves do not help predict their winning streaks, we will create a MultiIndex using `b_fighter` and `r_fighter` columns. 

```python
df.columns = df.columns.str.lower().str.strip()
df.columns = df.columns.str.replace('\(seconds\)','_seconds' )
df.columns = df.columns.str.replace('ko/tko', 'ko_tko' )
df.set_index(['r_fighter', 'b_fighter'], inplace=True)
df.sample(n=5, random_state=999)
```

However, we will keep this column intact for ease of plotting the dataset later on. It is only after that we will drop the column. On the other hand, `referee`, `date`, `location` features are not necessary for both plotting and predictive modelling. #### Boxplot of the Reach of a Winner Compared to that of a Loser

As the dataset contains information about two fighters in each observation, we have decided to create a column that holds the results of subtractions on two columns, namely `r_reach_cms` and `b_reach_cms`. We will box-plot the results to see whether the median is anything but 0. 

```python
df_copy = df.copy()
df_copy = df_copy[(df_copy['winner'] == 'Red') | (df_copy['winner'] == 'Blue')]
df_copy.reset_index(drop=True, inplace=True)

num_winners = df_copy.shape[0]
# Create an ndarray to store the differences between the reaches of winners and losers
np_reach_diff = np.empty(num_winners)

for index, row in df_copy.iterrows():
    if row['winner'] == 'Red':
        np_reach_diff[index] = row['r_reach_cms'] - row['b_reach_cms']
    elif row['winner'] == 'Blue':
        np_reach_diff[index] = row['b_reach_cms'] - row['r_reach_cms']

plt.figure(figsize= (16,8))
sns.histplot(np_reach_diff, kde=True).set_title('Figure 2: the Reach of a Winner Compared to That of a Loser', fontsize = 15)

plt.xlabel("Winner's reach $-$ Loser's reach")
plt.show()
```

As we can see above, a small difference in lengths of arms does not seem to play a significant role in the sport. There are slightly more win counts for those with a longer reach than those with a proportionally shorter reach. 

### Bivariate Visualisation

We shall now investigate what attributes correlate with fighters on a winning streak. We will assemble an ad hoc dataset on fighters currently (21/03/2021) on a winning streak for more than 2 consecutive fights. 

```python
r_column_names = df.columns[df.columns.str.startswith('r_')].tolist() + ['weight_class']
r_winners = df.loc[df['winner'] == 'Red', r_column_names]
r_winners.reset_index(level=1, drop=True, inplace=True)
r_winners.reset_index(level=0, inplace=True)
r_winners.rename(columns={name : name[2:] for name in r_column_names if name.startswith('r_')}, inplace=True)
r_winners.rename(columns={'r_fighter':'fighter'}, inplace=True)

b_column_names = df.columns[df.columns.str.startswith('b_')].tolist() + ['weight_class']
b_winners = df.loc[df['winner'] == 'Blue', b_column_names]
b_winners.reset_index(level=0, drop=True, inplace=True)
b_winners.reset_index(level=0, inplace=True)
b_winners.rename(columns={name : name[2:] for name in b_column_names if name.startswith('b_')}, inplace=True)
b_winners.rename(columns={'b_fighter':'fighter'}, inplace=True)

winners = pd.concat([r_winners, b_winners]).sort_index(kind='merge')
winners = winners.set_index('fighter')
winners = winners[~winners.index.duplicated(keep='first')]
winners_win_streak = winners[(winners['post_fight_win_streak'] > 2)]
```

From Figure 4A and 4C down below, we glean that in the absence of a preponderance of leg kick attempts, a conspicuous proportion of fighters still rack up many significant strikes. On the other hand, body shot attempts have a less weaker association with significant strikes, most prominently in the range of 0 to 10 shots on the x-axis, as compared to leg kick attempts. 

```python
avg_sig_str_landed = pd.concat([df.loc[df['winner'] == 'Red', 'r_avg_sig_str_landed'], df.loc[df['winner'] == 'Blue', 'b_avg_sig_str_landed']])
avg_body_att = pd.concat([df.loc[df['winner'] == 'Red', 'r_avg_body_att'], df.loc[df['winner'] == 'Blue', 'b_avg_body_att']])
avg_leg_att = pd.concat([df.loc[df['winner'] == 'Red', 'r_avg_leg_att'], df.loc[df['winner'] == 'Blue', 'b_avg_leg_att']])
        
plt.figure(figsize=(16,12))
plt.subplot(2,2,1)
sns.regplot(avg_body_att, avg_sig_str_landed)
plt.title('Figure 4A: Significant Strikes and Bodyshot Attempts')
plt.xlabel('Bodyshot Attempts')
plt.ylabel('Significant Strikes')

plt.subplot(2,2,2)
sns.regplot(avg_leg_att, avg_sig_str_landed, color='g')
plt.title('Figure 4B: Significant Strikes and Legkick Attempts')
plt.xlabel('Legkick Attempts')
plt.ylabel('Significant Strikes')

plt.show()
```

Women's weight divisions have more marked variability in the average number of landed headshots, indicating greater variability in technique compared to the men's.

```python
losing_b_fighters = df.loc[df['b_post_fight_lose_streak'] > 1, ['b_avg_opp_head_landed', 'b_age', 'weight_class']]
winning_b_fighters = df.loc[df['b_post_fight_win_streak'] > 1, ['b_avg_opp_head_landed', 'b_age', 'weight_class']]
losing_r_fighters = df.loc[df['r_post_fight_lose_streak'] > 1, ['r_avg_opp_head_landed', 'r_age', 'weight_class']]
winning_r_fighters = df.loc[df['r_post_fight_win_streak'] > 1, ['r_avg_opp_head_landed', 'r_age', 'weight_class']]

losing_b_fighters.reset_index(level=0, drop=True, inplace=True)
winning_b_fighters.reset_index(level=0, drop=True, inplace=True)
losing_b_fighters.rename(columns={'b_avg_opp_head_landed': 'avg_opp_head_landed', 'b_age':'age'},inplace=True)
winning_b_fighters.rename(columns={'b_avg_opp_head_landed': 'avg_opp_head_landed', 'b_age':'age'},inplace=True)
losing_b_fighters.index.name = 'fighter'
winning_b_fighters.index.name = 'fighter'

losing_r_fighters.reset_index(level=1, drop=True, inplace=True)
winning_r_fighters.reset_index(level=1, drop=True, inplace=True)
losing_r_fighters.rename(columns={'r_avg_opp_head_landed': 'avg_opp_head_landed', 'r_age':'age'},inplace=True)
winning_r_fighters.rename(columns={'r_avg_opp_head_landed': 'avg_opp_head_landed', 'r_age':'age'},inplace=True)
losing_r_fighters.index.name = 'fighter'
winning_r_fighters.index.name = 'fighter'

losing_fighters = pd.concat([losing_b_fighters, losing_r_fighters])
winning_fighters = pd.concat([winning_b_fighters, winning_r_fighters])
losing_fighters = losing_fighters[(losing_fighters['age'] >= 20) & (losing_fighters['age'] <= 50)]
winning_fighters = winning_fighters[(winning_fighters['age'] >= 20) & (winning_fighters['age'] <= 50)]
losing_fighters['age'] = pd.qcut(losing_fighters['age'], 3, labels=['20s', '30s', '40s'])
winning_fighters['age'] = pd.qcut(winning_fighters['age'], 3, labels=['20s', '30s', '40s'])

plt.figure(figsize=(16,8))
plt.subplot(1, 2, 1)
ax = sns.barplot(data=losing_fighters, x='age', y='avg_opp_head_landed', hue='weight_class', hue_order=['Flyweight', 'Bantamweight', 'Featherweight', 'Lightweight', 'Welterweight', 'Middleweight', 'LightHeavyweight', 'Heavyweight', 'WomenStrawweight', 'WomenFlyweight', 'WomenBantamweight', 'WomenFeatherweight'])
ax.set(xlabel='Age', ylabel='Average Number of Headshots Absorbed')
plt.ylim(0, 70)
plt.title('Figure 5A: Losers: Average Headshots Landed by Opponents')

plt.subplot(1, 2, 2)
ax = sns.barplot(data=winning_fighters, x='age', y='avg_opp_head_landed', hue='weight_class', hue_order=['Flyweight', 'Bantamweight', 'Featherweight', 'Lightweight', 'Welterweight', 'Middleweight', 'LightHeavyweight', 'Heavyweight', 'WomenStrawweight', 'WomenFlyweight', 'WomenBantamweight', 'WomenFeatherweight'])
ax.set(xlabel='Age', ylabel='Average Number of Headshots Absorbed')
plt.ylim(0, 70)
plt.title('Figure 5B: Winners: Average Headshots Landed by Opponents')

plt.show()
```

## Summary and Conclusion <a id='summary_and_conclusion'></a>

Our raw dataset was by no means perfect as it came with a few logical slips that needed correction, after which we were confident in the integrity of our derived response feature `R_post_fight_win_streak`. We have dropped the features whose values definitely do not contribute to the better fit of our predictive model. For Phase 2, we have renamed the columns as required and created another version of the dataset that contains no outliers. 

Through visualisation of our dataset, we now understand a merit of weight cutting, especially for those whose walking around weights indicate they belong to the lightweight division, but they trim down a great deal to the featherweight to gain a weight advantage. As fight counts in each division are indicative of its total population, there is a higher probability that the lightweight and welterweight divisions will house a larger number of highly technical fighters. For that reason, lightweight fighters who think like statisticians might want to consider competing at the featherweight class. However, those who weigh between the welterweight and middleweight divisions may want to skimp on weight-cutting and try the heavier division should they have an unsuccessful run in the lighter division.  

We intuitively understand fighters with a longer reach are at an advantage. Data have shown slightly higher win counts for slightly lengthier fighters. However, the difference in counts gets rather significant when there is 10 centimeters or more difference in the reach. 

We take for granted more weapons will serve well fighters in their pursuit of winning. Hence fighters today display spectacular kicks as part of their weaponry. One of the key factors determining who wins is the number of significant strikes. Data show that fighters who win a match without leg-kick attempts still land as many significant strikes as those with 10 or more leg-kick attempts.

Fighters on a losing streak are characterized by a cumulative average of 20 or more shots to the head in all their previous fights. In comparison to them, fighters on a winning streak tend to absorb fewer headshots. Fighters might benefit from practicing parrying, ducking and sliding away punches aimed at their heads followed by their own share of punches to the opponent's head to increase their chances of winning. We also have seen that age has a positive association with the number of head shots absorbed in most weight divisions. For fighters who want to retire as unscathed as they once cut their teeth, it may be wise to get in early and leave the sport as early. 

We also see that grappling submission is a highly favored way for fighters on a winning streak to finish their opponents. In weight divisions where the average takedown percentage is high, fighters might thrive by honing their responsive submission skills to cancel out the effectiveness of a takedown. In weight divisions where the average takedown percentage is low, fighters could improve their takedown skills to take advantage of other fighters who are used to standing up and hence, perhaps, less skilled in defending takedowns. 

The dataset has far too many descriptive features to find all the meaningful correlations with the response feature through plotting only. We will undertake such a mission in Phase 2 through multiple regression. Should its model fit turn out to be not as promising as we now hope, we shall try NN models to predict post-fight winning streaks of fighters.
