In Phase 1, the raw dataset underwent a cleaning process, which resulted in two formats of `DataFrame` (one with outliers present, the other without). In the process, we also created a dependent variable derived from two existing features (We naively decided to keep one of the independent features for Phase 2 only to learn to drop it since it is a good practice to drop all features used to derive a response feature). ### Report Overview <a id='report_overview'></a>

We start by outlining the methodology used for fitting our datasets (one with outliers, the other without) into multiple linear regression models; we take the approach of using two datasets to build respective models and selecting the one with outliers present if the conclusions are the same. 

After generating our full models, we will analyse the results and run diagnostics to check for any violation of model assumptions. We will then conduct feature selection to produce our reduced models, which will be also analysed and examined in the same manner as in the case of full models. 

We will then discuss the limitations of our models and critique their results. Next, we will summarise the entire project and its findings. Lastly we finish the report by drawing conclusions with respect to our goals and objectives.

### Overview of Methodology <a id='overview_of_methodology'></a>

Each of our datasets contains 134 features used to make predictions with respect to our dependent feature `r_post_fight_win_streak`. With this many "independent" features from observational data, our models run the risk of being influenced by multicollinearity. We will address this potential problem by removing features with the Variance Inflation Factors (VIFs) larger than 5. We are going to mitigate severe multicollinearity since our goal is both to make predictions and to understand the impact of each independent variable (Frost, n.d.). We then build our full models. After that, we run diagnostics to check whether the model assumptions are met. We then perform backward elimination with the p-value approach to select the predictors that help increase the strength of our model fit. We will then again confirm that the model conditions are met for the second time.

Now that we have assigned our model formulas to strings, we are ready to test for multicollinearity with VIFs. It takes rather cumbersome computing to filter out the features subject to multicollinearity. Below is the code used for the job: 


```python
def filter_out_high_vifs(fse, dfe, y, X): # fse == formula_string_encoded, dfe == df_encoded
    
    # Creating a dataframe that will contain the names of all the feature variables and their VIFs
    vif = pd.DataFrame()
    vif['VIF Factor'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    vif['features'] = X.columns
    vif = vif.sort_values(by = "VIF Factor", ascending = False)
    
    # Drop a feature with a VIF Factor of 5 or higher and repeat the process until there's no such features left.
    while True:
        if vif.iloc[0]['features'] == 'Intercept':
            if vif.iloc[1]['VIF Factor'] < 5 and vif.iloc[1]['VIF Factor'] != 'inf':
                del y, X, vif
                return fse
            else:
                feature_to_drop = vif.iloc[1]['features']
                vif_factor = vif.iloc[1]['VIF Factor']
                idx = vif.iloc[1].index.values
                
        else:
            if vif.iloc[0]['VIF Factor'] < 5 and vif.iloc[0]['VIF Factor'] != 'inf':
                del y, X, vif
                return fse
            else:
                feature_to_drop = vif.iloc[0]['features']
                vif_factor = vif.iloc[0]['VIF Factor']
                idx = vif.iloc[0].index.values
            
        print(f'{feature_to_drop} with the VIF of {vif_factor:.2f} has been dropped')
        dfe.drop(feature_to_drop, axis=1, inplace=True)
                 
        fse = fse.replace(' + ' + feature_to_drop, '')
        X.drop(columns=[feature_to_drop], inplace=True)
        vif = pd.DataFrame()
        vif['VIF Factor'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
        vif['features'] = X.columns
        vif = vif.sort_values(by = "VIF Factor", ascending = False)
    

y, X = patsy.dmatrices(formula_string_encoded, df_encoded, return_type='dataframe') 
formula_string_encoded = filter_out_high_vifs(formula_string_encoded, df_encoded, y, X)
y, X = patsy.dmatrices(formula_string_encoded_wo, df_encoded_wo, return_type='dataframe') 
formula_string_encoded_wo = filter_out_high_vifs(formula_string_encoded_wo, df_encoded_wo, y, X)
```

We will now see how precise predicted values are in relation to the actual values. 

````python
def plot_line(axis, slope, intercept, **kwargs):
        xmin, xmax = axis.get_xlim()
        axis.plot([xmin, xmax], [xmin*slope+intercept, xmax*slope+intercept], **kwargs)
        
fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_size_inches(16,8)

ax1.scatter(residuals_full['actual'], residuals_full['predicted'], alpha=0.3);
plot_line(axis=ax1,  slope=1, intercept=0, c='red')
ax1.set_xlabel('Actual Winning Streak')
ax1.set_ylabel('Predicted Winning Streak')
ax1.set_title('Figure 1A: Actual vs Predicted\nData with Outliers')

ax2.scatter(residuals_full_wo['actual'], residuals_full_wo['predicted'], alpha=0.3);
plot_line(axis=ax2,  slope=1, intercept=0, c='red')
ax2.set_xlabel('Actual Winning Streak')
ax2.set_ylabel('Predicted Winning Streak')
ax2.set_title('Figure 1B: Actual vs Predicted\nData Without Outliers')
plt.show()
````

Before we proceed to further interpret the regression outcomes, it would be sapient to check whether the regression assumptions have been met. 

### Full Model Diagnostic Checks

Now that we have our full models, we will check whether they meet the following four assumptions for multiple regression:

1. the residuals of the model are nearly normal,
2. the residuals against the fitted values plot illustrates constant variability,
3. the residuals against the time feature plot feature shows no pattern (independence),
4. each independent variable is linearly related to the outcome. 

#### Normal Distribution Condition
Let us first check the first assumption by plotting histograms of residuals and their Q-Q plots.

#### Constant Variability Condition
We shall check for homoscedasticity across absolute-residuals-against-fitted-values plots with LOWESS lines drawn through them.

````python
full_residuals = abs(residuals_full['residual'])
full_fitted = residuals_full['predicted']

full_residuals_wo = abs(residuals_full_wo['residual'])
full_fitted_wo = residuals_full_wo['predicted']

from statsmodels.nonparametric.smoothers_lowess import lowess
smoothed = lowess(full_residuals, full_fitted)
smoothed_wo = lowess(full_residuals_wo, full_fitted_wo)

fig, (ax1, ax2) = plt.subplots(1,2)
fig.set_size_inches(16,8)
ax1.scatter(full_fitted, full_residuals, edgecolors = 'k', facecolors = 'none')
ax1.plot(smoothed[:,0], smoothed[:,1], color = 'r')
ax1.set_title('Figure 4A: Versus Fits\nData with Outliers')
ax1.set_xlabel('Fitted Values')
ax1.set_ylabel('Absolute Value of Residuals')
ax1.plot([min(full_fitted), max(full_residuals)], [0,0], color = 'k', linestyle = ':', alpha = .3)

ax2.scatter(full_fitted_wo, full_residuals_wo, edgecolors = 'k', facecolors = 'none')
ax2.plot(smoothed_wo[:,0], smoothed_wo[:,1], color = 'r')
ax2.set_title('Figure 4B: Versus Fits\nData Without Outliers')
ax2.set_xlabel('Fitted Values')
ax2.set_ylabel('Absolute Value of Residuals')
ax2.plot([min(full_fitted_wo), max(full_residuals_wo)], [0,0], color = 'k', linestyle = ':', alpha = .3)

fig.tight_layout(pad=1.0)

plt.show()
````

We observe that the left plot from the outlier-model exhibits non-constant variability. In comparison to it, the dots from the without-outlier-model are scattered with more constant variability, but far from the ideal homoscedasticity with the lump in the middle of the lowess line in Figure 4B.

### Feature Selection

Our full models contained many features with p-values greater than the significance level of 0.05, indicating many of the terms are statistically insignificant. We now perform backward elimination on our full models with the p-value approach.

````python
def backward_feature_selection(fse, mff, dfe):
    ## create the patsy model description from formula
    patsy_description = patsy.ModelDesc.from_formula(fse)
    
    # initialize feature-selected fit to full model
    linreg_fit = mff
    
    # do backwards elimination using p-values
    p_val_cutoff = 0.05
    
    ## WARNING 1: The code below assumes that the Intercept term is present in the model.
    ## WARNING 2: It will work only with main effects and two-way interactions, if any.
    
    print('\nPerforming backwards feature selection using p-values:')
    
    while True:
        pval_series = linreg_fit.pvalues.drop(labels='Intercept')
        pval_series = pval_series.sort_values(ascending=False)
        term = pval_series.index[0]
        pval = pval_series[0]
        if (pval < p_val_cutoff):
            break
        term_components = term.split(':')
        print(f'\nRemoving term "{term}" with p-value {pval:.4}')
        if (len(term_components) == 1): ## this is a main effect term
            patsy_description.rhs_termlist.remove(patsy.Term([patsy.EvalFactor(term_components[0])]))    
        else: ## this is an interaction term
            patsy_description.rhs_termlist.remove(patsy.Term([patsy.EvalFactor(term_components[0]), 
                                                        patsy.EvalFactor(term_components[1])]))    
        linreg_fit = smf.ols(formula=patsy_description, data=dfe).fit()
        
    ## this is the clean fit after backwards elimination
    model_reduced_fitted = smf.ols(formula = patsy_description, data = dfe).fit()
    #########
    print("\n***")
    print(model_reduced_fitted.summary())
    print("***")
    print(f"Regression number of terms: {len(model_reduced_fitted.model.exog_names)}")
    print(f"Regression F-distribution p-value: {model_reduced_fitted.f_pvalue:.4f}")
    print(f"Regression R-squared: {model_reduced_fitted.rsquared:.4f}")
    print(f"Regression Adjusted R-squared: {model_reduced_fitted.rsquared_adj:.4f}")
    
    return model_reduced_fitted

model_reduced_fitted = backward_feature_selection(formula_string_encoded, model_full_fitted, df_encoded)
model_reduced_fitted_wo = backward_feature_selection(formula_string_encoded_wo, model_full_fitted_wo, df_encoded_wo)
````

## Critique & Limitations <a id='critique_and_limitation'></a>

All of our models, regardless of whether outliers are present and both full and reduced, overestimated the actual values when they should be 0; the predicted winning streaks also included negative values, at odds with the minimum value of a winning streak in real life: 0. They also underestimated for the majority of the time, if not all the time, when the actual winning streak is greater than 2. Since the response variable is discrete and is limited to non-negative values, a Poisson regression or one of its variants should instead have been used to fit data (Grace-Martin, n.d.).

The outliers-models also clearly exhibited non-constant variability, an indication that models other than linear regression may result in a better fit. The no-outliers-models were more constant in variability than their counterparts, but the graphs 4B and 9B still beg the question of whether different models should be tried for results that can be trusted. 

Despite our response variable being technically numerical as its value ranges from 0 to 7 and 0 to 16 for outliers-models and no-outliers-models resepctively, its underyling nature is categorical and numerical; the most ideal model would be able to predict whether a fighter wins or not as well as how many wins he's won prior to the fight. Given the complex natrue of our chosen response variable, we have failed to produce a linear model that performs reasonably well enough to be useful in predicting winning streaks.
